```python
import pandas as pd
```

```python
edges = pd.read_csv('./edges.csv')
edges.head()
```

```python
previews = pd.read_csv('./spotify-previews.csv')
previews = previews.loc[
    (previews['artists_coincides']) & (previews['track_name_conincides'])
]
previews.head()
```

```python
def getTrack(a1, a2):
    preview = previews.loc[(previews['source'] == a1) & (previews['target'] == a2), ['spotify_artists', 'preview']]
    if (len(preview.values) > 0):
        return preview.values[0]
    return ['', '']
```

```python
edges_records = []
for i, edge in edges.iterrows():
    track = getTrack(edge['source'], edge['target'])
    record = {
        "source": edge['source'],
        "target": edge['target'],
        "weight": edge['weight'],
        "track_name": edge['colab_track_name'],
        "artists": track[0],
        "preview": track[1],
    }
    edges_records.append(record)
```

```python
nodes = pd.read_csv('./nodes.csv')
nodes.head()
```

```python
images = pd.read_csv('./images.csv')
images.head()
```

```python
def getImage(id):
    image = images.loc[images['id'] == id]
    if (len(image.values) > 0):
        return image.values[0][1]
    return ''
```

```python
# Rename columns, merge with images
nodes_records = []
for i, node in nodes.iterrows():
    record = {
        "spotify_id": node['id.1'],
        "name_id": node['id'],
        "name": node['label'],
        "followers": node['followers'],
        "popularity": node['popularity'],
        "betweeness_centrality": node['betweenesscentrality'],
        "group": node['modularity_class'],
        "unique_collabs": node['unique_collabs'],
        "total_collabs": node['total_collabs'],
        "image": getImage(node['id.1'])
    }
    nodes_records.append(record)
```

```python
nodes_df = pd.DataFrame(nodes_records)
nodes_df.to_csv('./nodes-full.csv')
nodes_df.head()
```

```python
edges_df = pd.DataFrame(edges_records)
edges_df.to_csv('./edges-full.csv')
edges_df.head()
```
