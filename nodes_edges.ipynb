```python
import pandas as pd
import numpy as np
```

```python
edges = pd.read_csv('./input/edges.csv')
edges.head()
```

```python
previews = pd.read_csv('./input/spotify-previews.csv')
previews = previews.loc[
    (previews['artists_coincides']) & (previews['track_name_conincides'])
]
previews.head()
```

```python
nodes = pd.read_csv('./input/nodes.csv')
nodes.head()
```

```python
images = pd.read_csv('./images.csv')
images.head()
```

```python
def getTrack(a1, a2):
    preview = previews.loc[(previews['source'] == a1) & (previews['target'] == a2), ['spotify_artists', 'preview', 'id']]
    if (len(preview.values) > 0):
        return preview.values[0]
    return ['', '', '']
```

```python
def getArtistId(id_name):
    _id = nodes.loc[nodes['id'] == id_name, 'id.1']
    if (len(_id.values) > 0):
        return _id.values[0]
    return ""
```

```python
def getImage(id):
    image = images.loc[images['id'] == id]
    if (len(image.values) > 0):
        return image.values[0][1]
    return ''
```

```python
edges_records = []
for i, edge in edges.iterrows():
    track = getTrack(edge['source'], edge['target'])
    record = {
        "source": edge['source'],
        "target": edge['target'],
        "weight": edge['weight'],
        "track_name": edge['colab_track_name'],
        "artists": track[0],
        "preview": track[1],
        "source_id": getArtistId(edge['source']),
        "target_id": getArtistId(edge['target']),
        "track_id": track[2],
    }
    edges_records.append(record)
```

```python
# Rename columns, merge with images
nodes_records = []
for i, node in nodes.iterrows():
    record = {
        "spotify_id": node['id.1'],
        "name_id": node['id'],
        "name": node['label'],
        "followers": node['followers'],
        "popularity": node['popularity'],
        "betweeness_centrality": node['betweenesscentrality'],
        "group": node['modularity_class'],
        "unique_collabs": node['unique_collabs'],
        "total_collabs": node['total_collabs'],
        "image": getImage(node['id.1'])
    }
    nodes_records.append(record)
```

```python
nodes_df = pd.DataFrame(nodes_records)
nodes_df.to_csv('./output/nodes-full.csv')
nodes_df.head()
```

```python
edges_df = pd.DataFrame(edges_records)
edges_df.to_csv('./output/edges-full.csv')
edges_df.head()
```

```python
n_nodes = len(nodes)
n_edges = len(edges)
n_images = len(images)
n_preview = len(previews)

print(f"{n_nodes} nodos/artists")
print(f"{n_images} images/profile pic de artista")

print(f"{n_edges} edges/colaboraciones")
print(f"{n_preview} preview/url de 30s de track")
```

```python
scrollyArtistsIds = ["Shakira", "Rauw Alejandro", "Bad Bunny", "Nicky Jam"]
scrollyEdges = []
scrollyArtists = []

for i, row in nodes_df.iterrows():
    if (row['name_id'] in scrollyArtistsIds):
        scrollyArtists.append(row)

for i, row in edges_df.iterrows():
    if (
        row['source'] in scrollyArtistsIds and
        row['target'] in scrollyArtistsIds
    ):
        scrollyEdges.append(row)


pd.DataFrame(scrollyArtists).to_csv('./output/scrollyArtists.csv')
pd.DataFrame(scrollyEdges).to_csv('./output/scrollyEdges.csv')
```
